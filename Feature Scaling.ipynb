```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel (r"C:\Users\Admin\Machine Learning\loan.xlsx")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["Age"]])
```

```python
ss.transform(dataset[["Age"]])
```

```python
dataset["Age"]=pd.DataFrame(ss.transform(dataset[["Age"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Age"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Age"])
plt.show()
```

## Feature Scalling(Normalization)

```python
dataset.describe()
```

```python
sns.distplot(dataset["Salary"])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["Salary"]])
```

```python
dataset["Salary_min"]=ms.transform(dataset[["Salary"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Salary"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Salary_min"])
plt.show()
```
