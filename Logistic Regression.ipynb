```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel(r"C:\Users\Admin\Machine Learning\houseprice_2.xlsx")
dataset.drop(columns=["bedrooms"], inplace=True)
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="price",y="parking", data=dataset)
plt.show()
```

```python
x=dataset[["price"]]
y=dataset["parking"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=52)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[13300000]])
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="price",y="parking", data=dataset)
sns.lineplot(x="price",y=lr.predict(x),data=dataset,color="red")
plt.show()
```
