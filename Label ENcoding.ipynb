# Ordinal Encoding

```python
df2 =pd.DataFrame({"Size":["s", "m","l","xl","l","m","s","s","l","xl","m"]})
df2.head(10)
```

```python
ord_data =[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(df2[["Size"]])
```

```python
df2["Size_en"]= oe.transform(df2[["Size"]])
```

```python
df2
```

```python
# Map Function
```

```python
ord_data1={"s":0,"m":1, "l":2,"xl":3}
```

```python
df2["Size_en_map"]=df2["Size"].map(ord_data1)
```
